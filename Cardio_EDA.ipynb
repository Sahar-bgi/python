```python
import pandas as pd
df=pd.read_csv("/content/CardioVascular.csv")
df.head()
```

```python
#Fetch the name of the Columns into a list
col=df.columns.to_list()
col
```

```python
#Fetch each columns' values into a seperate list
sbp=[df['sbp(mm Hg)'].to_list()]
ldl=[df['ldl(mmol/L)'].to_list()]
adip=[df['adiposity'].to_list()]
tob=[df['tobacco(in Kg)'].to_list()]
alc=[df['alcohol'].to_list()]
typa=[df['typea'].to_list()]
age=[df['age'].to_list()]
fh=[df['FamHist'].to_list()]
hd=[df['CHD'].to_list()]
sbp
```

```python
#Creating a Dictionary
dic={
col[0]:sbp,
col[1]:ldl,
col[2]:adip,
col[3]:tob,
col[4]:alc,
col[5]:typa,
col[6]:age,
col[7]:fh,
col[8]:hd
}
```

```python
df=pd.DataFrame(dic)
df
```

```python
#Save Dataset into Excel
df.to_excel('\cardio.xlsx')
```
